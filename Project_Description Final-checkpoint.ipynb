Here, we'll work with magnetic resonance (MR) imaging data from the Sunnybrook Cardiac Dataset. The full image is a 3D time series spanning a single heartbeat. These data are used by radiologists to measure the ejection fraction: the proportion of blood ejected from the left ventricle during each stroke.

```python
import imageio
import numpy as np
import scipy.ndimage as ndi
import matplotlib.pyplot as plt
                                 
```

### Let's import one sample 

You may refer to next source to download sample with number SCD2001_MR_117:

https://www.kaggle.com/datasets/salikhussaini49/sunnybrook-cardiac-mri

```python
# YOUR CODE HERE
imag = imageio.imread("SCD2001_MR_117.dcm")
print(f'Image type: {type(imag)}')
print(f'Shape of image array: {imag.shape}')
```

### Show the image

```python
# YOUR CODE HERE
plt.imshow(imag)
```

### Make the plot with grayscale

```python
# YOUR CODE HERE
plt.imshow(imag, cmap='gray') 
```

### Make plot the colorbar 

```python
# YOUR CODE HERE
imag = imag.astype(np.float64)
print(f'Data type: {imag.dtype}')
print(f'Min. value: {imag.min()}')
print(f'Max value: {imag.max()}')

# Plot the grayscale image
plt.imshow(imag, cmap='gray', vmin=0, vmax=255) #increase the contrast 
plt.colorbar()
```

### Rotation

Rotate the figure and show both of them

```python
# YOUR CODE HERE
imag_Rotate = ndi.rotate(imag, angle=-30, reshape=False)
# Plot the original and rotated images
fig, axes = plt.subplots(1, 2)
axes[0].imshow(imag,cmap='gray')
axes[1].imshow(imag_Rotate,cmap='gray')
```

### Histogram

Plot the histogram as shown below 

```python
# YOUR CODE HERE
hist = ndi.histogram(imag, min=0, max=255, bins=25)
# Plot the histogram and CDF
plt.plot(hist)
```

### Mask

Create next mask

```python
# YOUR CODE HERE
mask_imag = (imag>=70) & (imag<150)
plt.imshow(mask_imag, cmap='gray')
```

### Show selected part

Show the part selected by using previous mask

```python
# YOUR CODE HERE
imag_part = np.where(mask_imag, imag, 0)
plt.imshow(imag_part,cmap='gray')
```

### Tune a mask

Tune the previous maske as follows

```python
# YOUR CODE HERE
mask_bone = imag>=150
mask_dilate = ndi.binary_dilation (mask_bone, iterations=5)  #binary_dilation: Add pixels along edges
mask_closed = ndi.binary_closing (mask_bone, iterations=5)  #binary_closing: Dilate then erode, "filling in" holes
# Plot masked images
fig, axes = plt.subplots(1,3)
axes[0].imshow(mask_bone)
axes[1].imshow(mask_dilate)
axes[2].imshow(mask_closed)
```

### Show result of running mask

```python
# YOUR CODE HERE
imagMask = np.where(mask_closed, imag, 0)
plt.imshow(imagMask,cmap='gray')
```

### Edge detection

```python
# YOUR CODE HERE
weights = [[+1, 0, -1], [+1, 0, -1], [+1, 0, -1]] 
edges = ndi.convolve(imag, weights)
plt.imshow(edges, cmap='seismic', vmin=-150, vmax=150)
plt.colorbar()
```

## Segmentation

### Identify labels

```python
# YOUR CODE HERE
im_filt = ndi.median_filter(imag, size=3)

# Select high-intensity pixels
mask_start = np.where(im_filt>60, 1, 0)
mask = ndi.binary_closing(mask_start)

# Label the objects in "mask"
labels, nlabels = ndi.label(mask)
print('Num. Labels:',nlabels)

# Create a `labels` overlay
overlay = np.where(labels>0, labels, np.nan)

# Use imshow to plot the overlay
plt.imshow(overlay, cmap='rainbow', alpha=0.75)
```

### Select objects

```python
# YOUR CODE HERE
labels, nlabels = ndi.label(mask)

# Select left ventricle pixels
lv_val = labels[128, 128]
lv_mask = np.where(labels == lv_val, 1, np.nan)

# Overlay selected label
plt.imshow(lv_mask, cmap='rainbow')
plt.show()
```

### Extract specific object

```python
# YOUR CODE HERE
labels, nlabels = ndi.label(mask)
lv_val = labels[128, 128]
lv_mask = np.where(labels == lv_val, 1, 0)

# Find bounding box of left ventricle
bboxes = ndi.find_objects(lv_mask) #reate a new image containing only the left ventricle.
print('Number of objects:', len(bboxes))
print('Indices for first box:', bboxes[0])

# Crop to the left ventricle (index 0)
im_lv = imag[bboxes[0]]

# Plot the cropped image
plt.imshow(im_lv,cmap='gray')
```
